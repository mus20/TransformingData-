```python
import pandas as pd
import numpy as np

df = pd.read_csv("hn_stories.csv",header=None, names=['a','b','c','d'])
df.head()
```

```python
df = df.dropna()
words = []
i=0
for header in df['d']:
    i=i+1
    splits = header.split(" ")
    for split in splits:
        split = ''.join(e for e in split if e.isalnum())
        words.append(split.lower())
print(words)
```

```python
unique = set(words)
print(len(unique))
print(len(words))
count_unique={}
for i in unique:
    count_unique[i] = words.count(i)
    
print(count_unique)        
```

```python
count_df = pd.DataFrame([count_unique])
```

```python
count_df = count_df.transpose()
count_df.sort_values(0,ascending=False)
```
